```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as ply
%matplotlib inline
```

```python
# Import Data
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header=None)
```

```python
# Data preprocessing
# Create a list of transactions as thay is accepted format for this model, 
# also make sure they are converted to strings

transactions =[]
for i in range(1, len(dataset)):
  transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
# print(transactions)
```

```python
# min_support
# we want atleast 3 transaction with items occuring, 7 is days(can be month etc etc depending on dataset) of dataset collected
# and support is 3*7 /7501 = 0.0027, we take 0.003 as min_support
# round((3*7 / len(dataset)*1000))
```

```python
# min_confidence 
# use rule of thumbs, default value = 0.8 (got no rules), it was too high, try 0.8/2 see how many rules
# and how relevant they are, else again 0.4/2 and then see how many rules we get...
```

```python
# min_lift, measure the quality of rules
# generally a good lift would be at least 3 and above.. based on experience
# lifts below 3 make the rules not that relevant
```

```python
"""
min_length and max_length:

We want to identify the best deals of buy one product A and get another product B for free.
And therefore the rules we want to get in the end must have only two products,
one product in the left hand side of the rule, 
and one product in the right hand side of the rule,
All right and therefore, to make sure we have this, you know
one product A on the left and one product B on the right,
well, we need to add two more arguments here
which is first min length, Okay, and then max length,
where of course min length is the minimum number 
of elements you want to have in your rule.
You know, left to right and max length is the maximum number
of elements you wanna have in your rule left to right.

In other words,
where of course min length is the minimum number
of elements you want to have in your rule.
You know, left to right and max length is the maximum number
of elements you wanna have in your rule left to right.
Also, depends on business requirements
Buy 2, get 2nd free, 
Buy 3 get 3rd free,
Buy 2, get 11th free, set min_length =2 and max_length = 11, 
* Please re check 
"""
```

```python
# Training the Apriori model on the dataset
from apyori import apriori
rules = apriori(transactions= transactions,
                min_support = 0.003,
                min_confidence=0.2,
                min_lift=3,
                min_length = 2,
                max_length = 2)
```

# Visualizing the results

```python
# Displaying the first results coming directly from the output of the apriori function
results = list(rules)
results
```

```python
results[0]
```

```python
# Putting the results well organized into a Pandas DataFrame 
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side',
                                                               'Right Hand Side',
                                                               'Support',
                                                               'Confidence',
                                                               'Lift'])
```

```python
# Displaying the results non sorted
resultsinDataFrame
```
