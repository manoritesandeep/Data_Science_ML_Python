```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing dataset - Ad Click Through Rate - Optimizing the click thru rate
dataset = pd.read_csv('Ads_CTR_Optimisation.csv')
```

```python
import random

# Number of users - len(df.axes[0]) or len(df)
N = len(dataset.axes[0])
print(f"Total number of users are: {N}")

# Number of ads we have - len(df.axes[1])
d = len(dataset.axes[1])
print(f"Total number of Ads are: {d}")

# List of selected Ads over the round
ads_selected = []

# the number of times the ad i got reward 1 up to round n 
number_of_rewards_1 = [0]*d

# the number of times the ad i got reward 0 up to round n 
number_of_rewards_0 = [0]*d
total_reward = 0
for n in range(0, N):
  ad = 0
  max_reward = 0
  for ads in range(0,d):
    random_beta = random.betavariate(number_of_rewards_1[ads] + 1,
                                     number_of_rewards_0[ads] + 1)
    if random_beta > max_reward:
      max_reward = random_beta
      ad = ads

  ads_selected.append(ad)
  reward = dataset.values[n, ad]
  if reward == 1:
    number_of_rewards_1[ad] += 1
  else:
    number_of_rewards_0[ad] += 1    
  total_reward += reward
```

```python
print(ads_selected)
```

```python
print(total_reward)
```

```python
# Visualizing the results
# We would pick the machine with the highest Confidence Bound

plt.hist(ads_selected)
plt.title("Histogram of ads selections")
plt.xlabel("Ads")
plt.ylabel("Number of times each ad was selected")
plt.show()
```
